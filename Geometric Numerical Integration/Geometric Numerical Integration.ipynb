```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
```

## Examples and Numerical Experiments

$$
\frac{\partial u}{\partial t} = u(\alpha - \beta v) \\
\frac{\partial v}{\partial t} = v(\delta u - \gamma)
$$

```python
def lotka_volterra(u, v, alpha, beta, delta, gamma):
    r"""
    du/dt = u * (alpha - beta * v)
    dv/dt = v * (delta * u - gamma)
    """
    return u * (alpha - beta * v), v * (delta * u - gamma)
```

```python
x = np.arange(0, 5, 0.5)
y = np.arange(0, 5, 0.5)
x, y = np.meshgrid(x, y)
u, v = lotka_volterra(x, y, 1, 1, 1, 2)

plt.figure()
plt.title("Lotka-Volterra model vector field")
plt.xlabel("prey")
plt.ylabel("predator")
plt.quiver(x, y, u, v)
plt.show()
```

```python
y = []
h = 0.12
u, v = 2, 2
for i in range(100):
    du, dv = lotka_volterra(u, v, 1, 1, 1, 2)
    u = u + h * du
    v = v + h * dv
    y.append((u, v))
U, V = zip(*y)
plt.plot(U, V, "o-")
plt.title("Explicit Euler model for Lotka-Volterra model")
plt.xlabel("prey")
plt.ylabel("predator")
plt.show()
```
