**QUESTION 1.**

The prime factors of 13195 are 5, 7, 13 and 29. 

What is the largest prime factor of the number 600851475143?

```python
def max_prime_factors(x):
    prime_factors = set()
    i = 1
    while x > 1:
        while x % i == 0 and i > 1:
            x = x // i  # this allows the algorithm to be O(logN)
            prime_factors = prime_factors | {i}
        i += 1
    return max(prime_factors) if prime_factors else 1.0
```

```python
print(max_prime_factors(600851475143))
```

```python
def measure(test_func, x, repeat=1):
    start = time.time()
    for _ in range(repeat):
        test_func(x)
    return time.time() - start
```

```python
xs = np.arange(1, 1000000, 10000)
ys = np.array([measure(max_prime_factors, x) for x in xs])

plt.title("Performace measurement")
plt.xlabel("input number")
plt.ylabel("duration (sec.)")
plt.plot(xs, ys)
plt.show()
```
